```python
import pandas as pd
```

```python
df = pd.read_csv('maribelajar-steming.csv')

df
```

```python
df.drop(df.columns[[0]], axis = 1, inplace = True)
df
```

```python
df.dropna()
```

```python
from deep_translator import GoogleTranslator

def Translate(text):
    text = GoogleTranslator(source='id', target='en').translate(text)
    return text
```

```python
 print(df.isnull().sum())
```

```python
df.dropna(subset=["Steming"], axis=0, inplace=True)
```

```python
 print(df.isnull().sum())
```

```python
df['Text_Inggris'] = df['Steming'].apply(Translate)
```

```python
df
```

```python
keep_col = ['Steming','Text_Inggris']
df1 = df[keep_col]
df1.to_csv('maribelajar-translate.csv')
```
